```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from keras.layers import Activation, Dense, Flatten
import tensorflow as tf
from keras.models import Sequential
from keras import activations
from tensorflow.keras.utils import to_categorical
from sklearn import preprocessing
import tensorflow as tf

train_x = y[:8000]
test_x= y[8000:]
# сохранение значений для проверки
proverka_x=test_x

train_y = train_x[:,[3]]
train_y=train_y.astype(int)

test_y = test_x[:,[3]]
test_y=test_y.astype(int)
proverka_y=test_y

train_x = np.delete(train_x,3,1)
test_x = np.delete(test_x,3,1)

print('Тренировончная выборка')
print(train_x)
print('Ответы')
print(train_y)
print()
print('Тестовая выборка')
print(test_x)
print('Ответы')
print(test_y)

#преобразование тренировчной данных в скаляр (для облегчения обучения)
scaler = preprocessing.StandardScaler().fit(train_x)
train_x=scaler.transform(train_x)

scaler1 = preprocessing.StandardScaler().fit(test_x)
test_x=scaler1.transform(test_x)

```

Создание обучающей модели

```python
model = keras.Sequential([
    Dense(3,input_dim=3, activation='relu'),
    Dense(2, activation='relu'),
    Dense(3, activation='softmax')
])
print(model.summary())
```

Преобразование выходных данных в векторы по категориям

```python
train_y = to_categorical(train_y,3)
test_y = to_categorical(test_y,3)

print(train_x.shape)
train_y.shape
```

```python
model.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])
```

```python

history=model.fit(train_x, train_y, batch_size=15,epochs=15)
```

```python
model.evaluate(test_x,test_y)
```

Визуализация обучения нейронной сети

```python
plt.plot(history.history['accuracy'])
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend
plt.title("Graph of accuracy ")
plt.grid()
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend
plt.title("Graph of loss ")
plt.grid()
plt.show()
```
